## Import Packages

```python
import pandas as pd
from google.colab import files
```

## Load Data

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/Harumnesia/Machine-Learning/refs/heads/main/Dataset/Dataset%20Awal/dataset%20-%20PERFUME%20LOKAL.csv',
    dtype={'price': str}
)
df.head()
```

## Data Understanding

## Data Preparation

```python
df['price'] = df['price'].str.replace('.', '', regex=False)
df['price'] = df['price'].str.replace(',', '', regex=False)
df['price'] = df['price'].str.strip()
```

```python
df['price'] = df['price'].astype(int)
```

```python
df['gender'] = df['gender'].str.capitalize()
```

```python
df['top notes'] = df['top notes'].str.title()
df['mid notes'] = df['mid notes'].str.title()
df['base notes'] = df['base notes'].str.title()
```

## Save Dataset

```python
df.to_csv('Dataset_Harumnesia_clean.csv', index=False)
files.download('Dataset_Harumnesia_clean.csv')
```
